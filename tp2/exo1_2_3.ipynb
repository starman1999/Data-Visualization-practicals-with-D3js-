```python
import pandas as pd # Pandas
import numpy as np # Numpy
import matplotlib.pyplot as plt # Matplotlibrary
import seaborn as sns # Seaborn Library


productivity = pd.read_csv("../data/us-productivity.csv")
cities = pd.read_csv("../data/us-cities-top-1k.csv")
#print(productivity)
populations = []

for state in productivity['state']:
    population = 0
    for i in range(len(cities["State"])):
        if (state == cities["State"][i]):
            population += cities["Population"][i]
    populations.append(population)

print(populations)

sns.lineplot(x=populations, y= productivity["value"])
```
